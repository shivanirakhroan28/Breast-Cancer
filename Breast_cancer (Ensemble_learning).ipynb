# Separating the Features and Target

```python
x= pd.DataFrame(df.drop("Target",axis=1))
y= pd.Series(df['Target'])
```

```python
columns= x.columns
```

```python
columns
```

# Using MinMaxScaler To scale down values

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mn= MinMaxScaler()
```

```python
x= mn.fit_transform(x)
```

```python
x= pd.DataFrame(x,columns=columns)
```

```python
x
```

# splitting the data into training data & testing data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,random_state=42,test_size=0.2,shuffle=True,stratify=y)
```

```python
x_test
```

```python
# Base learning models
```

# Implementing model-1: KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn= KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
print("training score:", knn.score(x_train,y_train))
print("testing score:", knn.score(x_test,y_test))
```

# Visualisation to check most perfect value  for n_neighbors

```python
train_score=[]
test_score=[]

for i in range(1,15):
    knn= KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    train_score.append(knn.score(x_train,y_train))
    test_score.append(knn.score(x_test,y_test))
```

```python
rg= list(range(1,15))
plt.figure(figsize=(11,4))
plt.plot(rg,train_score,label='Train_score')
plt.plot(rg,test_score,label='Test_score')
plt.legend()
plt.show()
```

```python
knn= KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
print("training score:", knn.score(x_train,y_train))
print("testing score:", knn.score(x_test,y_test))
```

```python
knn_prob= knn.predict_proba(x_test)
```

```python
knn_prob=  knn_prob[:,1]
```

```python
result= pd.DataFrame(knn_prob,columns=['KNN_PROB'])
```

```python
result
```

```python
pred= knn.predict(x_test)
```

```python
pred
```

# Implementing model-2: DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc= DecisionTreeClassifier(random_state=42,criterion='gini') 
```

```python
dtc.fit(x_train,y_train)
```

```python
print("Training score:", dtc.score(x_train,y_train))
print("Testing score:", dtc.score(x_test,y_test))
```

```python
dtc_prob= dtc.predict_proba(x_test)[:,1]
```

```python
result["DTC_PROB"]= dtc_prob
```

```python
result
```

# Implementing model-3: LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg=LogisticRegression() 
```

```python
lg.fit(x_train,y_train)
```

```python
print("Training score:", lg.score(x_train,y_train))
print("Testing score:", lg.score(x_test,y_test))
```

```python
lg_prob= lg.predict_proba(x_test)[:,1]
```

```python
result['Lg_PROB']= lg_prob
```

```python
result
```

```python
result['Proba_AVG']= (result['KNN_PROB']+result["DTC_PROB"]+result['Lg_PROB'])/3
```

```python
result
```

```python
vr=[]
for i in result['Proba_AVG']:
    if i<0.5:
        vr.append(0)
    else:
        vr.append(1)
```

```python
result['Output']=vr
```

```python
result
```

```python
result['Actual_value']= np.array(y_test)
```

```python
result
```
